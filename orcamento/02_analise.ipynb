### Estudando os atendimentos

```python
caminho = '/content/drive/My Drive/Datascience/orcamento/dataset/tratado/'

def le_planilha(planilha_dados, separador = ",", decimais = ",", encode='utf-8'):
  df_dados = pd.read_csv(caminho + planilha_dados, decimal=decimais, sep=separador, encoding=encode)
  return df_dados
```

```python
def qtd_atendimentos_anteriores(cod_paciente, data_internacao):
  df_temp = df_atendimentos[df_atendimentos.CD_PACIENTE == cod_paciente]
  return df_temp[df_temp.DH_ATENDIMENTO_INT < data_internacao].count()[0]
```

```python
def cria_planilha(dataframe, nome, caminho):
  dataframe.to_csv(caminho + nome)
```

###Transformando as colunas DATA em Pandas Datetime

```python
def converte_para_datetime(dataset, colunas):
  for coluna in colunas:
    dataset[coluna] = pd.to_datetime(dataset[coluna])
```

```python
colunas_data = ['DH_ATENDIMENTO_INT', 'DH_ALTA_INT', 'DH_ATENDIMENTO_URG', 'DH_ALTA_URG']
converte_para_datetime(df_atendimentos, colunas_data)
```
